```python
#Select the range of data to test
data_mult_2 = data_mult.loc[(data_mult['DateTime'] > '2026-03-16 02:48:00') & (data_mult['DateTime'] < '2026-03-16 03:00:00')]
```

```python
#Create a matrix to represent the x-axis of a graph
#x1 = np.arange(0, len(y), 1)
x2 = np.arange(0, len(data_mult_2["ID"]), 1)
x2
```

```python
#Select Electrical Grid Voltage variable
y2_vg = data_mult_2[["Voltaje Red Eléctrica"]]
print(y2_vg)
```

```python
#Electrical Grid Voltage vs Time
plt.figure(figsize=(24,10))
plt.title("Electrical Grid Voltage vs Time", fontsize = 16)
plt.scatter(x2, y2_vg, s=5, color="red", label="Electrical Grid Voltage")
plt.plot(x2, y2_vg, lw=0.8, color="blue", label="Electrical Grid Voltage")
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('Time', fontsize = 16)
plt.ylabel('Electrical Grid Voltage', fontsize = 16)
plt.legend(fontsize = 16)
plt.grid()
plt.show()
```

```python
#Select Inverter Relay Voltage variable
y2_vr = data_mult_2[["Voltaje Relay Inversor"]]
print(y2_vr)
```

```python
#Inverter Relay Voltage vs Time
plt.figure(figsize=(24,10))
plt.title("Inverter Relay Voltage vs Time", fontsize = 16)
plt.scatter(x2, y2_vr, s=5, color="red", label="Inverter Relay Voltage")
plt.plot(x2, y2_vr, lw=0.8, color="blue", label="Inverter Relay Voltage")
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('Time', fontsize = 16)
plt.ylabel('Inverter Relay Voltage', fontsize = 16)
plt.legend(fontsize = 16)
plt.grid()
plt.show()
```

```python
#Select Battery Voltage variable
y2_vb = data_mult_2[["Voltaje Batería"]]
print(y2_vb)
```

```python
#Battery Voltage vs Time
plt.figure(figsize=(24,10))
plt.title("Battery Voltage vs Time", fontsize = 16)
plt.scatter(x2, y2_vb, s=5, color="red", label="Battery Voltage")
plt.plot(x2, y2_vb, lw=0.8, color="blue", label="Battery Voltage")
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('Time', fontsize = 16)
plt.ylabel('Battery Voltage', fontsize = 16)
plt.legend(fontsize = 16)
plt.grid()
plt.show()
```

```python
#Select Voltage to Power On Inverter variable
y2_vi = data_mult_2[["Voltaje Encender Inversor"]]
print(y2_vi)
```
